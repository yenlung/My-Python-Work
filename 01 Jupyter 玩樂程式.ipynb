我們要學的畫圖方式主要只有兩種:
    
* `plt.plot`: 畫折線圖
* `plt.scatter`: 畫散佈圖

首先是最重要、最常用的 `plt.plot`。

```python
plt.plot([2, -5, 3, 8, -2])
```

我們把 `plt.plot` 引入一個串列 (list) 的數字, 串列都是用中括號。這裡輸入的是:

    [2, -5, 3, 8, -2]

其實是點的 Y 座標。標準的 `plt.plot` 指令是

    plt.plot(X, Y)

其中 X 是點的 X 座標的串列, Y 是點的 Y 座標的串列。

如果我們沒有給 X 座標, X 座標會自動以 0, 1, 2, ... 補上。然後 `plt.plot(X, Y)` 就是把這些點點出來, 再用線段連接起來, 成為折線圖。

```python
# Python 的註解就是像這樣先用井字號
# 這樣 Python 就不會理會我們打什麼
# 接下來我們想把 (0, 2.1), (1.1, 3.4), (2.5, 1), (3.2, 5), (4.8, 7) 這幾個點連起來!
# 之後我們會學更美的呈現法!
```

```python
X = [0, 1.1, 2.5, 3.2, 4.8]
Y = [2.1, 3.4, 1, 5, 7]
```

```python
plt.plot(X, Y)
```

所以 `plt.plot` 標準用法就是把 

$$(x_1, y_1), (x_2, y_2), \ldots (x_n, y_n)$$

幾個點點上, 用線段連起來 (也就是折線圖)。指令的下法是把 x, y 座標分別收集起來:

$$x = [x_1, x_2, \ldots, x_n]$$

$$y = [y_1, y_2, \ldots, y_n]$$

然後用

    plot(x,y)
    
畫出來。

我們怎麼畫圖呢? 基本上也是描點, 也就是用 `plt.plot`。

![極座標](images/polar.png)

我們可以用 Jupyter 秘技, 很方便打入希臘字母! 最炫的是, 這些希臘字母可以當變數名稱!

![latex+tab](images/latex_tab.png)

現在我們來畫個圖!

```python
π = np.pi
```

```python
θ = np.linspace(0, 2*π, 500)
```

```python
r = 3
x = r * np.cos(θ)
y = r * np.sin(θ)
```

```python
plt.plot(x, y)
```

這根本不圓啊!! 我們要來調整一下。

這時, 需要知道 `matplotlib` 畫圖的基本架構。

![matplotlib 圖的結構](images/figure_axes.png)

我們要設定一組 axes 的屬性, 就需要知道目前在用的 axes 是哪一組。這要怎麼知道呢? 就是用

    ax = plt.gca()
    
`gca` 的意思是 "Get Current Axes"。

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
```

果然成功了!

附帶一提, 我們之所以用了 `r=3` 這樣的設定, 是因為如此很容易造出「可變半徑」的極座標系統。比方說, 我們想畫出:

$$r = 1 - \sin(\theta),$$

可以這樣畫出來...

```python
r = 1 - np.sin(θ)

x = r * np.cos(θ)
y = r * np.sin(θ)
```

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y)
```

但「藍心」看來怪怪的, `matplotlib` 有可「快速上色法」, 用一個字母就代表一個顏色。

![matplotlib 快速顏色](images/matplotlib_quick_color.png)

```python
ax = plt.gca()
ax.set_aspect('equal')

plt.plot(x, y, 'r')
```
